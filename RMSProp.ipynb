#   محمد عادل الغنام RMSProp 

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = random.sample(range(0, 20), 20)
a = -1
b = 2
y=[]

for i in range(len(x)):
    hx = (-1*x[i])+2
    y.append(hx)
    
print(x)
print(y)
```

```python
x=np.array(x)
y=np.array(y)
```

```python
th0=0
th1=0
y_pred = []
loss_func = []
th0_list = []
th1_list = []
Epochs = []
v0 = 0
v1 = 0
alpha = 0.1
e = 1*10**-8
B = 0.9


for i in range(2000):
    Epochs.append(i)
    hx = th0 + th1*x
    y_pred.append(hx)
    cost_func = (1/(2*len(x)))*sum((hx-y)**2)
    loss_func.append(cost_func)
    g0 = (1/len(x))*sum((hx-y))
    g1 = (1/len(x))*sum(((hx-y)*x))
    v0 = (B*v0) + ((1-B)*(g0**2))
    v1 = (B*v1) + ((1-B)*(g1**2))
    th0 = th0 - ((alpha/((v0**0.5)+e))*g0)
    th0_list.append(th0)
    th1 = th1 - ((alpha/((v1**0.5)+e))*g1)
    th1_list.append(th1)
```

```python
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.scatter(Epochs , loss_func )
plt.plot(Epochs , loss_func)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
plt.scatter(th0_list , loss_func)
plt.plot(th0_list , loss_func)
plt.xlabel("theta 0")
plt.ylabel("Loss")
plt.show()
```

```python
plt.scatter(th1_list , loss_func)
plt.plot(th1_list , loss_func)
plt.xlabel("theta 1")
plt.ylabel("Loss")
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
for i in range(2000):
    plt.plot(x , th0_list[i] + (th1_list[i]*x) )
```

```python
plt.scatter(x,y)
plt.plot(x , th0 + (th1*x))
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
y_p = th0 + (th1*x)
ss_residual = sum((y-y_p)**2)
ss_total = sum((y-np.mean(y))**2)
r2 = 1-(ss_residual/ss_total)
print(r2)
```
