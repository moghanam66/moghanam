Stochastic

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = np.array([1400,1600,1700,1875,1100,1550,2350,2450,1425,1700])

price = np.array([245000,312000,279000,308000,199000,219000,405000,324000,319000,255000])

```

```python
plt.scatter(size , price)
plt.show()
```

```python
th0 = 150000
th1 = 44.5
loss_func = np.zeros(len(size))
lr= 0.000001

for i in range(len(size)):
    hx = th0 + (size[i]*th1)
    th0 = th0 - (lr*(hx-price[i]))
    th1 = th1 - (lr*(hx-price[i])*size[i])
    loss_func[i] = (0.5*(hx-price[i]))
    
    
print(th0)
print(th1)
    
    
```

```python
print(loss_func)
```

```python
z = np.arange(len(size))
plt.plot(z,loss_func)
plt.show()
```

```python
price_pred = th0 + (th1*size)
```

```python
plt.scatter(size , price)
plt.plot(size , price_pred)
plt.show()
```
