# محمد عادل الغنام ADAM

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = random.sample(range(0, 20), 20)
a = -1
b = 2
y=[]

for i in range(len(x)):
    hx = (-1*x[i])+2
    y.append(hx)
    
print(x)
print(y)
```

```python
x=np.array(x)
y=np.array(y)
```

```python
th0=0
th1=0
y_pred = []
loss_func = []
th0_list = []
th1_list = []
Epochs = []
v0 = 0
v1 = 0
alpha = 0.01
e = 1*10**-8
B1 = 0.9
B2 = 0.999
m0 = 0
m1 = 0

for i in range(1,2000):
    Epochs.append(i)
    hx = th0 + th1*x
    y_pred.append(hx)
    cost_func = (1/(2*len(x)))*sum((hx-y)**2)
    loss_func.append(cost_func)
    g0 = (1/len(x))*sum((hx-y))
    g1 = (1/len(x))*sum(((hx-y)*x))
    m0 = (B1*m0) + ((1-B1)*g0)
    m1 = (B1*m1) + ((1-B1)*g1)
    v0 = (B2*v0) + ((1-B2)*(g0**2))
    v1 = (B2*v1) + ((1-B2)*(g1**2))
    m0_ = m0/(1-(B1**i))
    m1_ = m1/(1-(B1**i))
    v0_ = v0/(1-(B2**i))
    v1_ = v1/((1-B2**i))
    th0 = th0 - ((alpha*m0_)/((v0_**0.5)+e))
    th0_list.append(th0)
    th1 = th1 - ((alpha*m1_)/((v1_**0.5)+e))
    th1_list.append(th1)
    
```

```python
print(th0)
print(th1)
```

```python
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.scatter(Epochs , loss_func )
plt.plot(Epochs , loss_func)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
plt.scatter(th0_list , loss_func)
plt.plot(th0_list , loss_func)
plt.xlabel("theta 0")
plt.ylabel("Loss")
plt.show()
```

```python
plt.scatter(th1_list , loss_func)
plt.plot(th1_list , loss_func)
plt.xlabel("theta 1")
plt.ylabel("Loss")
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
for i in range(len(th0_list)):
    plt.plot(x , th0_list[i] + (th1_list[i]*x) )
```

```python
plt.scatter(x,y)
plt.plot(x , th0 + (th1*x))
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
